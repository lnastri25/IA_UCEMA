```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

# Generar los datos XOR
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

# Crear y entrenar la red neuronal
mlp = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=10000, random_state=42)
mlp.fit(X, y)

# Crear una malla de puntos para graficar la frontera de decisión
xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 500),
                     np.linspace(-0.5, 1.5, 500))
Z = mlp.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Graficar la frontera de decisión y los puntos XOR
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.viridis)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.viridis, edgecolor='k', s=100)
plt.title("XOR entrenado con una red neuronal usando sklearn")
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.show()
```
