```python
# Comenzamos con un gráfico básico de lineas
import matplotlib.pyplot as plt
a = [1,2,8,9]
b = [1,20,10,5]
plt.plot(a, b)
plt.show()
```

```python
# Cambiamos el tipo de línea
plt.plot(a, b, color='red', linewidth=3)
plt.show()
```

```python
# Agregamos una leyenda 
plt.plot(a, b, color='red', linewidth=3, label="Compras")
plt.legend()
plt.show()
```

```python
# Graficamos desde un dataframe de pandas 
import numpy as np
import pandas as pd
d = np.array([1,20,10,5])
df = pd.DataFrame(data=d, columns=['Compras'], index=[1,2,8,9])
df.plot()
plt.show()
```

```python
# Graficamos desde un csv 
df = pd.read_csv('/content/compras.data', usecols=['id','compras'], 
                 index_col = 0)
df.plot()
plt.show()
```

```python
# Agregamos grilla, título y nombre de ejes
df.plot()
plt.title('Compras en cada Q')
plt.xlabel('Id. del cliente')
plt.ylabel('Compras del cliente')
plt.grid()
plt.show()
```

```python
# Agregamos otro gráfico y cambiamos el ancho de la línea de uno de ellos
df1 = pd.read_csv('/content/comprasQ1.data', usecols=['id','q1'], 
                 index_col = 0)
df2 = pd.read_csv('/content/comprasQ2.data', usecols=['id','q2'], 
                 index_col = 0)
ax = df1.plot(linewidth=5)
df2.plot(ax=ax)
plt.title('Compras en cada Q')
plt.xlabel('Id. del cliente')
plt.ylabel('Compras del cliente')
plt.grid()
plt.show()
```

```python
# Gráfico de barras
df = pd.read_csv('/content/estad_historicas.data', 
                 usecols=['Año','Ventas'], 
                 index_col = 0)
df.plot.bar(color='lightblue')
plt.title('Ventas Históricas')
plt.ylabel('Ventas ($MM)')
plt.show()
```

```python
# Más de una columna en un gráfico de barras
df = pd.read_csv('/content/estad_historicas.data', 
                 index_col = 0)
df.plot.bar()
plt.title('Estadísticas Históricas')
plt.show()
```

```python
# Histograma
n = np.array([1,2,3,1,4,1,1,2,1,5,1,3,1,3,3])
df = pd.DataFrame(n)
df.plot.hist()
plt.title('Cantidad de compras por cliente')
plt.show()
```

```python
# Agrupando varios en una única categoría
n = np.array([10,20,30,15,40,16,13,24,11,52,12,31,10,30,38])
df = pd.DataFrame(n)
df.plot.hist(bins=[10,20,30,40,50])
plt.title('Cantidad de compras por cliente')
plt.show()
```

```python
# Agregando separación entre barras
df.plot.hist(bins=[10,20,30,40,50], rwidth=0.8, color='lightcoral')
plt.title('Cantidad de compras por cliente')
plt.xlabel('Compras')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Eliminar leyenda
df.plot.hist(bins=[10,20,30,40,50], rwidth=0.8, color='lightcoral').legend('')
plt.title('Cantidad de compras por cliente')
plt.xlabel('Compras')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de dispersión
df = pd.read_csv('/content/estad_historicas.data',index_col = 0)
df.plot.scatter(0,1)
plt.show()
```

```python
# Gráfico de dispersión con más de una categoría
df = pd.read_csv('/content/estad_hist2.data',index_col = 0)
colores = {'auto':'navy','castillo':'limegreen'}
coloresTipo = df.Tipo.map(colores)
fig, ax = plt.subplots()
ax.scatter(df.Compras,df.Ventas,color=coloresTipo)
plt.show()
```

```python
# Agregar leyenda al gráfico de dispersión (con más de una categoría) 
# y modificar tamaño de los puntos 

df = pd.read_csv('/content/estad_hist2.data',index_col = 0)
colores = {'auto':'navy','castillo':'limegreen'}
coloresTipo = df.Tipo.map(colores)
fig, ax = plt.subplots()
for tipo in set(df.Tipo):
    ax.scatter(
        df.Compras[df.Tipo == tipo],
        df.Ventas[df.Tipo == tipo],
        s = 80,
        c = colores[tipo],
        label = tipo)
plt.legend()
plt.show()
```

```python
# Gráfico de torta
df = pd.read_csv('/content/estad_historicas.data', 
                 index_col = 0)
df.plot.pie(y='Ventas')
plt.title('Estadísticas de Ventas')
plt.show()
```

```python
# Mejorando el formato del gráfico de torta
df.plot.pie(y='Ventas',
            shadow=True,
            explode=(0.15,0,0,0,0.2,0),
            legend=False)
plt.title('Estadísticas de Ventas')
plt.show()
```
